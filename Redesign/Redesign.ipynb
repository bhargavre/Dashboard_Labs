```python
import pandas
```

```python
trial = pandas.read_excel("Data on the global flow of people_Version March2014.xlsx")
```

```python
del trial['region_orig_id']
del trial['region_dest_id']
del trial['metadata']
```

```python
trial1=trial
```

```python
for index, row in trial1.iterrows():
    if row['country_orig_id'] == row['country_dest_id']:
        trial1.drop(index, inplace=True)
```
